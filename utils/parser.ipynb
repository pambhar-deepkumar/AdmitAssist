```python

class ModuleParser:
    def __init__(self):
        # Configure Docling pipeline with OCR and table structure support
        self.doc_converter = DocumentConverter(
            format_options={
                "PDF": PdfFormatOption(
                    pipeline_options={
                        "do_ocr": True,
                        "do_table_structure": True,
                        "table_structure_options": {
                            "do_cell_matching": True
                        }
                    }
                )
            }
        )
    
    def extract_module_info(self, pdf_path):
        # Convert PDF to structured format
        result = self.doc_converter.convert(pdf_path)
        
        module_data = {
            "code": "",
            "title": "",
            "credits": 0,
            "prerequisites": [],
            "content": "",
            "learning_outcomes": "",
            "teaching_language": "",
            "assessment": ""
        }
        
        # Extract structured information using layout analysis
        for section in result.sections:
            if "Prerequisites" in section.title:
                module_data["prerequisites"] = self._parse_prerequisites(section.text)
            elif "Contents" in section.title:
                module_data["content"] = section.text
            elif "Qualification Aims" in section.title:
                module_data["learning_outcomes"] = section.text
                
        return module_data
```
